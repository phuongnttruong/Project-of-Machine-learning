```python
G = osm.to_graph(nodes, edges, graph_type="networkx")
```

```python

import numpy as np    # library for numerical computations (vectors, matrices, tensors)
import pandas as pd    # library for data manipulation and analysis
import matplotlib.pyplot as plt    # library providing tools for plotting data

import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('orders_autumn_2020.csv')
data = data[['TIMESTAMP','USER_LAT', 'USER_LONG', 'VENUE_LAT', 'VENUE_LONG']]
```

```python
source_nodes = []
for x, y  in zip(data['USER_LONG'], data['USER_LAT']):
    source_nodes.append(ox.distance.nearest_nodes(G, x,y))
```

```python
targets_nodes = []
for x, y  in zip(data['VENUE_LONG'], data['VENUE_LAT']):
    targets_nodes.append(ox.distance.nearest_nodes(G, x,y))
```

```python
source_nodes[1]
```

```python
targets_nodes[1]
```

```python
routes =[]
```

```python
import networkx as nx
```

```python
for x, y  in zip(source_nodes, targets_nodes):
    routes.append(nx.shortest_path_length(G, x,y,  weight="length"))
```

```python
data['length'] = pd.Series(routes)
```
